**버블정렬(bubble sort)**

```python
def bubbleSort(A):
  # print(A)
  n = len(A)
  for last in range(n-1, 0, -1):
    for i in range(0, last):
      if A[i] > A[i+1]:
        A[i], A[i+1] = A[i+1], A[i]
    #   print(":", A)      # 회차별 결과 확인용
    # print(last, ":", A)  # 회차별 결과 확인용
  return A  #최종결과 반환용

A = [3,7,2,5]
bubbleSort(A)
```
