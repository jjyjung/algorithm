```python
import copy
def topological_sort1(W):
  E = copy.deepcopy(W)		# 깊은 복사
  A = []                  # 노드 기록
  for i in range(len(E.keys())):
    u = find_start_node(E).pop()
    A.append(V[u])
    E.pop(u)
  return A

def find_start_node(E):
  F = set(E.keys())
  for k in E.keys():
    for v in E[k]:
      if v in F:
        F.remove(v)
  return F

print(topological_sort1(E))
```
